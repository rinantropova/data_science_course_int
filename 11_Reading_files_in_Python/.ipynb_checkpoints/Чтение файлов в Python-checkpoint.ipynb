```python
with open('anna_words.txt', encoding='UTF-8') as file:
    print(file.readline())
```

```python
words = []
with open('anna_words.txt', encoding='UTF-8') as file:
    for word in file:
        words.append(word)
        
print(len(words))
```

```python
with open('anna_words.txt',encoding='UTF-8') as file:
    words = file.readlines()

print(len(words))
```

```python
words_set = set()
with open('anna_words.txt',encoding='UTF-8') as file:
    for word in file:
        words_set.add(word)

print(len(words_set))
        
```
